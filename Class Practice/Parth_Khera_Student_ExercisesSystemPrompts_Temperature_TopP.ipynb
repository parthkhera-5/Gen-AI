````python
from transformers import pipeline

# Load the chat model (same model we used in class)
pipe = pipeline('text-generation', model='TinyLlama/TinyLlama-1.1B-Chat-v1.0')

# Helper function (from class)
def get_response(messages, temperature=0.7, max_new_tokens=100):
    prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    outputs = pipe(prompt, max_new_tokens=max_new_tokens, do_sample=True, temperature=temperature)
    return outputs[0]["generated_text"].split("<|assistant|>")[-1]

print("Setup complete! ✅")
````

### Exercise 2.2: The Repeat Test — Deterministic or Random? 🎲

**Task:** Run the same prompt **3 times at the same temperature**.
- First, 3 runs at `temperature = 0.1`
- Then, 3 runs at `temperature = 1.2`

Observe: in which case are the outputs more similar to each other?


````python
prompt = "The most important invention in history is"

# TODO: loop 3 times at temperature 0.1
for i in range(3):
    res = generator(prompt, max_length=30, do_sample=True, temperature=0.1)  # <-- fix
    print(f"Run {i+1}:", res[0]['generated_text'])
    print("-" * 60)
````

````python
# TODO: now 3 times at temperature 1.2
for i in range(3):
    res = generator(prompt, max_length=30, do_sample=True, temperature=1.2)  # <-- fix
    print(f"Run {i+1}:", res[0]['generated_text'])
    print("-" * 60)
````

**🤔 Reflection Question 2.2:** Why were the outputs similar at low temperature? Explain in terms of the probability distribution (Hint: remember the "Serious Student" analogy — it only picks the highest-probability word).




At low temperature, the outputs were very similar because the model strongly prefers the highest-probability words. The probability distribution becomes more concentrated, so the same words are selected repeatedly. At high temperature, lower-probability words also have a chance of being selected, resulting in more varied and creative outputs.

## Part 3: Top-P (Nucleus Sampling) — The "Selection Pool" 🎯

Remember:
- **Top-P = 0.1** → the AI only looks at the top 1-2 words (probability sum = 10%)
- **Top-P = 0.9** → a bigger pool, more creativity, but higher chance of mistakes


### Exercise 3.1: Top-P Comparison

**Task:** Same prompt, different top_p values: `0.1`, `0.5`, `0.95`

Prompt: `"In a distant galaxy, humans discovered a planet made of"`


````python
prompt = "In a distant galaxy, humans discovered a planet made of"

# TODO: top_p = 0.1
res = generator(prompt, max_length=50, do_sample=True, top_p=0.1)  # <-- fix
print("TOP-P 0.1:\n", res[0]['generated_text'])
````

````python
# TODO: top_p = 0.5
res = generator(prompt, max_length=50, do_sample=True, top_p=0.5)  # <-- fix
print("TOP-P 0.5:\n", res[0]['generated_text'])
````

````python
# TODO: top_p = 0.95
res = generator(prompt, max_length=50, do_sample=True, top_p=0.95)  # <-- fix
print("TOP-P 0.95:\n", res[0]['generated_text'])
````

**🤔 Reflection Question 3.1:** Temperature and Top-P — both control randomness, but the **mechanism is different**. Explain the difference between the two in your own words:


Temperature changes the probability of every possible word by making the distribution sharper or flatter. Top-P works differently by selecting only the smallest group of words whose total probability reaches the chosen value (such as 0.9), then randomly choosing from that group. Temperature changes the shape of the distribution, while Top-P limits the pool of words from which the model can choose.
