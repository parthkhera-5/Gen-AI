## Question 1

Explain the difference between Encoder-only and Decoder-only architectures with a real-world example not discussed in class.

**Answer:**

**Encoder-only** architecture is designed to understand and analyze the input text. It reads the entire input at once and creates a meaningful representation (embedding) of the text. It does not generate new text. Since it focuses on understanding the input, it is mainly used for tasks such as sentiment analysis, text classification, spam detection, named entity recognition (NER), and semantic search.


**Decoder-only** architecture is designed to generate text. It predicts one word (or token) at a time based on the words that have already been generated. It uses masked (causal) self-attention, which prevents it from looking at future words while generating the current word. This makes it suitable for tasks such as chatbots, story generation, code generation, text completion, and question answering.




**Real-World Example**
Consider a hospital management system.
Suppose a doctor writes the following patient note:

"The patient has had a high fever for three days, a persistent cough, and difficulty breathing."

An encoder-only model reads the entire note, understands the symptoms, and identifies that the patient may have a respiratory infection. It can classify the urgency level, extract important medical terms, or recommend the appropriate medical department. Its role is only to analyze and understand the information; it does not create new text.

A decoder-only model, on the other hand, can use the doctor's note as a prompt to generate a discharge summary, explain the condition in simple language for the patient, or draft a follow-up instruction such as:

"Based on your symptoms, please continue taking the prescribed medication, stay hydrated, monitor your temperature regularly, and return to the hospital immediately if your breathing becomes worse."

Here, the decoder is creating new text from the given context.

## Question 2

Why is bidirectional context useful for text understanding? Give two NLP tasks where it is beneficial.

**Answer:**

Bidirectional context means that a model considers both the previous words and the following words in a sentence while understanding the meaning of each word. This helps the model capture the complete context of the sentence, leading to a more accurate interpretation.

Example:
1) Please turn on the light.
2) The bag is very light.

Here in the sentence 1 context refers to the electric light, but in the sentence 2 context refers to the weight(light).

Tasks of NLP:
1. Sentiment Analysis
In sentiment analysis, the model determines whether a sentence expresses a positive, negative, or neutral opinion.
Example: "The movie was not good."
The word "not" changes the meaning of "good" from positive to negative. Bidirectional context helps the model correctly classify the sentiment as negative.

2. Named Entity Recognition (NER)
NER identifies and classifies entities such as person names, organizations, locations, and dates. Understanding the surrounding words helps the model correctly identify the type of entity.
Example: "Apple launched a new iPhone."
Here, "Apple" refers to the company, not the fruit, because of the surrounding words "launched" and "iPhone".



## Question 3

Why can't a Decoder see future tokens during training? Explain with an example.

**Answer:**

A decoder cannot see future tokens because it is designed to predict the next token using only the tokens that have already appeared. This is called autoregressive generation. To achieve this, the decoder uses masked self-attention, which hides all future tokens during training. If the decoder could see future words, it would simply copy the answer instead of learning how to predict it. This would make the model ineffective during real-world text generation, where future words are not available.

**Example:**
Suppose the sentence is: "I love playing cricket."

During training, the decoder generates one word at a time.

To predict "love", it can only see: "I"

To predict "playing", it can only see: "I love"

To predict "cricket", it can only see: "I love playing"

It cannot see the future words while predicting the current word.

**What happens if token are visible?**
Suppose the decoder is predicting the word "playing".
If it could already see: "I love playing cricket"

then it would immediately know that the next word is "playing". The model would not actually learn how to predict the next word from context—it would simply copy the answer. This is known as data leakage or information leakage.

## Question 7

What is Masked Self-Attention? Explain why it prevents cheating.

**Answer:**

Masked Self-Attention (also called Causal Self-Attention) is an attention mechanism used in which each token can attend only to itself and the tokens before it. It cannot attend to future tokens because they are hidden (masked).

This prevents cheating because during training the entire target sequence is fed in at once for parallel efficiency — without the mask, each position could attend to (and effectively "see") the correct answer it's supposed to be predicting, making training trivial and useless for real generation, where future tokens simply don't exist at inference time.

## Question 8

Differentiate Parameters, Weights, and Biases with examples.

**Answer:**

**Parameters:** Parameters are the learnable values of a machine learning or deep learning model. They are automatically updated during training to minimize the prediction error.

Example: If a model has 3 weights and 1 bias, then the total number of parameters = 4.

**Weights:** Weights determine the importance or influence of each input feature on the output. They are learned during training.

Example: In predicting house prices, the weight tells how much the house size affects the price.

**Biases:** The learnable offset added after the weighted sum (the b in y = Wx + b); lets the model shift output independent of input.

Example: output = w1 * x1 + w2 * x2 + b → w1, w2 are weights, b is the bias, {w1, w2, b} together are parameters.

## Question 9

If a model has 15 billion parameters stored in FP16, estimate the memory required.

**Answer:**

Memory = Number of Parameters × Bytes per Parameter

Given:

Number of parameters = 15 billion = 15 × 10^9

Data type = FP16 (16-bit floating point)

FP16 stores each parameter in 16 bits = 2 bytes.

Calculation:

15 × 10^9 × 2 bytes = 30 × 10^9 bytes = 30GB

## Question 17

Interview Q1: Why do larger parameter counts not always produce better performance?

**Answer:**

Larger parameter counts do not always produce better performance because performance depends on data quality, training process, and model architecture, not just the number of parameters.

**Reasons:**
1. Overfitting: Large models may memorize the training data instead of learning general patterns, reducing performance on new data.
2. Data Quality: If the training data is limited or poor in quality, additional parameters provide little or no benefit.
3. Higher Computational Cost: Larger models require more memory, GPUs, and training time.
4. Diminishing Returns: Beyond a certain point, increasing parameters leads to only small improvements while greatly increasing cost.
5. Slower Inference: Large models take longer to generate predictions, making them less suitable for real-time applications.

## Question 18

Interview Q2: What happens if model weights are randomly initialized and never trained?

**Answer:**

If a model's weights are randomly initialized and never trained, the model will produce random and inaccurate predictions because it has not learned any patterns from the training data.

Effects:
1. Makes random predictions.
2. Does not learn relationships between inputs and outputs.
3. Produces very low accuracy.
4. Cannot generalize to new data.

```python
from transformers import pipeline
import time
```

```python
# Load BERT Fill-Mask Pipeline
fill_mask = pipeline(
    "fill-mask",
    model="bert-base-uncased"
)

text = "The capital of France is [MASK]."

start = time.time()

result = fill_mask(text)

end = time.time()

print("BERT Predictions:\n")

for i in result:
    print("Word :", i["token_str"])
    print("Score:", round(i["score"],4))
    print("Sentence:", i["sequence"])
    print()

bert_time = end-start
```

```python
generator = pipeline(
    "text-generation",
    model="gpt2"
)

prompt = "Artificial Intelligence is"

start = time.time()

output = generator(
    prompt,
    max_length=50,
    num_return_sequences=1
)

end = time.time()

print(output[0]["generated_text"])

gpt_time = end-start
```

```python
print("BERT Runtime :", bert_time,"seconds")
print("GPT Runtime :", gpt_time,"seconds")
```
